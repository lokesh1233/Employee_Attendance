```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


import warnings
import itertools
warnings.filterwarnings('ignore')
plt.style.use('fivethirtyeight')
# import statsmodels.api as sm
from fbprophet import Prophet
import matplotlib

matplotlib.rcParams['axes.labelsize'] = 14
matplotlib.rcParams['xtick.labelsize'] = 12
matplotlib.rcParams['ytick.labelsize'] = 12
matplotlib.rcParams['text.color'] = 'k'
```

```python
attendance_df = pd.read_csv("../data/processed/AttendaenceProcess.csv", parse_dates=True).set_index(['Date'])
attendance_df.index = attendance_df.index.astype('datetime64[ns]')
```

```python
sampleData = attendance_df.resample("W-MON").sum()
sampleData.plot(figsize=(15, 6))
plt.show()
```

```python
attendance_df.reset_index(inplace=True)
attendance_df.columns = ["ds", "y"]
```

```python
# Python
m = Prophet(yearly_seasonality=True, daily_seasonality=True)
m.fit(attendance_df)
```

```python
# Python
future = m.make_future_dataframe(periods=90)
future.tail()
```

```python
# Python
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
# Python
fig1 = m.plot(forecast)
```

```python
# Python
fig2 = m.plot_components(forecast)
```

### growth

```python
attendance_df['cap'] = 50
m = Prophet(growth='logistic', yearly_seasonality=True, daily_seasonality=True)
m.fit(attendance_df)
```

```python
# Python
future = m.make_future_dataframe(periods=10)
future['cap'] = 50
future['floor'] = 0
fcst = m.predict(future)
fig = m.plot(fcst)
```

```python
# Python
# attendance_df['y'] = 25 - attendance_df['y']
attendance_df['cap'] = 50
attendance_df['floor'] = 0
future['cap'] = 50
future['floor'] = 0
m = Prophet(growth='logistic', yearly_seasonality=True, daily_seasonality=True, changepoint_prior_scale=0.15)
m.fit(attendance_df)
fcst = m.predict(future)
fig = m.plot(fcst)
```
