```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


import warnings
import itertools
warnings.filterwarnings('ignore')
plt.style.use('fivethirtyeight')
import statsmodels.api as sm
import matplotlib

matplotlib.rcParams['axes.labelsize'] = 14
matplotlib.rcParams['xtick.labelsize'] = 12
matplotlib.rcParams['ytick.labelsize'] = 12
matplotlib.rcParams['text.color'] = 'k'
```

```python
octo = pd.read_excel('../data/raw/October 2018.xls')
nov = pd.read_excel('../data/raw/November 2018.xls')
dec = pd.read_excel('../data/raw/December 2018.xls')
```

```python
dataset = pd.concat([octo, nov, dec], sort=False)
dataset.dropna(axis=0, how='all', thresh=10, inplace =True)
dataset.fillna("00:00", inplace =True)
```

```python
dataset.head(10)
```

```python
# def presentEmployess(empTime):
# #     print(empTime.values)
#     p31 = np.asarray(empTime.values)
#     return (p31 != '00:00').sum()
# #     return list(filter(lambda x: 1 if (int(x.split(":")[0]) > 0 or int(x.split(":")[1]) > 0) else 0, empTime.values))
    

meltData = pd.melt(dataset, value_vars=dataset.columns.values[1:], id_vars=['Days'])
meltData.set_index("Days", inplace=True)
meltData.drop(["Out Time", "Duration", "OT", "T Duration"], inplace=True)
meltData.reset_index(inplace=True)
meltData["Attendance"] = meltData["value"].apply(lambda x: 1 if (int(x.split(":")[0]) > 0 or int(x.split(":")[1]) > 0) else 0)
meltData.drop(["Days", "value"], axis=1, inplace=True)
meltData = meltData.groupby("variable").sum().reset_index()
meltData.columns = ["Date", "Attendance"]
meltData["Date"] = meltData["Date"].apply(lambda x: x.replace(year=2018))
meltData.set_index(["Date"], inplace=True)
```

```python
meltData.head(10)
```

```python
meltData.to_csv("../data/processed/AttendaenceProcess.csv")
```

### Building and processing data for the regression model

```python
reg_data = meltData.reset_index()

reg_data["Year"] = reg_data["Date"].apply(lambda x: x.year)
reg_data["Month"] = reg_data["Date"].apply(lambda x: x.month)
reg_data["Day"] = reg_data["Date"].apply(lambda x: x.day)
reg_data["is_NWeekoff"] = reg_data["Date"].apply(lambda x: 1 if x.weekday() < 5 else 0 )
#  2, 19 oct, 7,8,9 Nov 24, 25 dec
holiday_list = ["2018-10-2", "2018-10-19", "2018-11-7", "2018-11-8", "2018-12-9", "2018-12-25"]
reg_data["is_workingday"] = reg_data["Date"].apply(lambda x: 1 if (str(x.year)+"-"+str(x.month)+"-"+str(x.day) not in holiday_list and x.weekday() < 5)  else 0)
#reg_data["is_holiday"] = 
```

```python
process_reg = reg_data.drop(["Date"], axis=1)
```

```python
process_reg.to_csv("../data/processed/AttendaenceProcess_reg.csv", index=False)
```

```python
reg_data["Mon"] = reg_data["Date"].apply(lambda x: 1 if x.weekday() == 0 else 0)
reg_data["Tue"] = reg_data["Date"].apply(lambda x: 1 if x.weekday() == 1 else 0)
reg_data["Wed"] = reg_data["Date"].apply(lambda x: 1 if x.weekday() == 2 else 0)
reg_data["Thu"] = reg_data["Date"].apply(lambda x: 1 if x.weekday() == 3 else 0)
reg_data["Fri"] = reg_data["Date"].apply(lambda x: 1 if x.weekday() == 4 else 0)
reg_data["Sat"] = reg_data["Date"].apply(lambda x: 1 if x.weekday() == 5 else 0)
reg_data["Sun"] = reg_data["Date"].apply(lambda x: 1 if x.weekday() == 6 else 0)
```
